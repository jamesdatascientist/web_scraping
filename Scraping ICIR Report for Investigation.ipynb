#### Time to Start Scraping! In the below code, we will:
+ Download the web page containing the forecast.
+ Create a BeautifulSoup class to parse the page.
+ Find the div with id seven-day-forecast, and assign to seven_day
+ Inside seven_day, find each individual forecast item.
+ Extract and print the first forecast item.

```python
# importing the necessary packages
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.icirnigeria.org/category/investigations/")
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="tdi_123")
forecast_items = seven_day.find_all(class_="tdb_module_loop")
overNight = forecast_items[0]
print(overNight.prettify())
```

```python
base_url = 'https://www.icirnigeria.org/category/investigations'
get_title_list = []
get_author_list = []
get_date_list = []
get_url_list = []

int_count = 1

for link in base_url_list:
    page = requests.get(link)
    
    soup = BeautifulSoup(page.content, 'html.parser')
    seven_day = soup.find(id="tdi_123")
    
    title_tags = seven_day.select(".tdb_module_loop .td-image-wrap")
    title = [pt["title"] for pt in title_tags]
    
    author_tags = seven_day.select(".tdb_module_loop .td-post-author-name")
    author_list = [pt.get_text() for pt in author_tags]
    
    date_tags = seven_day.select(".tdb_module_loop .td-post-date")
    date_list = [pt.get_text() for pt in date_tags]
    
    url_tags = seven_day.select(".tdb_module_loop .td-image-wrap")
    url_list = [pt["href"] for pt in url_tags]

    get_title_list.append(title)
    get_author_list.append(author_list)
    get_date_list.append(date_list)
    get_url_list.append(url_list)
    
print(get_title_list)
print()
print(get_author_list)
print()
print(get_date_list)
print()
print(get_url_list)
```

```python
import pandas as pd

investigation_2022_report = pd.DataFrame({
    "Author": get_author_list,
    "Title": get_title_list,
    "Date": get_date_list,
    "URL":get_url_list
})

investigation_2022_report
```

### Data Cleaning

```python
from cleantext import clean

def clean_more(txt):
    clean_txt = clean(text=str(txt),
            fix_unicode=True,
            normalize_whitespace=True,
            strip_lines=True,
            to_ascii=True,
            lower=False,
            no_line_breaks=True,
            no_urls=True,
            no_emails=True,
            no_phone_numbers=True,
            no_numbers=False,
            no_digits=False,
            no_currency_symbols=False,
            no_punct=False,
            replace_with_punct="",
            replace_with_url="",
            replace_with_email="",
            replace_with_phone_number="",
            replace_with_number="123",
            replace_with_digit="0",
            replace_with_currency_symbol="$",
            lang="en"
            )
    return clean_txt
```

```python
investigation_2022_report['Author'] = investigation_2022_report['Author'].apply(clean_more)
investigation_2022_report['Author']
```

```python
# Remove html tag elements, &nbsp, n&nbsp, \r, \n, \xa0
import re

def removeHTMLTags(t):
    clean = re.compile('<.*?>')
    return re.sub(clean,'',str(t))

def remove_nbsp(t):
    clean = re.compile('n&nbsp;')
    return re.sub(clean,'',str(t))

def remove_nnbsp(t):
    clean = re.compile('&nbsp;') 
    return re.sub(clean,'',str(t))

def remove_newline(t):
    clean = re.compile('\n')
    return re.sub(clean,' ',str(t))

def remove_carriageReturn(t):
    clean = re.compile('\r')
    return re.sub(clean,' ',str(t))

def remove_xa0(t):
    return t.replace(u'\xa0', u' ')

def remove_hyphen(t):
    clean = re.compile('-')
    return re.sub(clean,' ',str(t))
```

```python
investigation_2022_report['Author'] = investigation_2022_report['Author'].apply(remove_hyphen)
investigation_2022_report['Author']
```
